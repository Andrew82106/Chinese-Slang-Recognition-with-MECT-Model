```python
import tqdm
def is_chinese(char):
    if '\u4e00' <= char <= '\u9fff':
        return True
    else:
        return False

def has_nbsp(text):
    if text == u'&nbsp;':
        return False
    return '\u00A0' in text


def is_valid_utf8(text):
    try:
        text.encode('utf-8').decode('utf-8')
        return True
    except UnicodeDecodeError:
        return False

with open("./wiki.txt", "r", encoding='utf-8') as f:
    cont = f.read()
    cont = cont.replace(" ", "").replace("\n", "")
suc = 0
fai = 0
with open("./wiki.bmes", "w", encoding='utf-8') as f:
    for i in tqdm.tqdm(cont):
        if len(i) != 1 or (not is_valid_utf8(i)) or (has_nbsp(i)):
            fai += 1
            continue
        if i not in ".。,，" and (not is_chinese(i)):
            fai += 1
            continue
        f.write(f"{i} O\n")
        if i == '。':
            f.write(f"\n")
        suc += 1
print(f"rate={suc/(suc+fai)}")
```
